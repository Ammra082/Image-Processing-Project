```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
from scipy import ndimage
import os
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# ----- LOAD AND PREPROCESS CIFAR-10 DATASET -----
print("Loading CIFAR-10 dataset...")
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
# Class names for CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# Convert labels from integers to one-hot encoding
train_labels_one_hot = tf.keras.utils.to_categorical(train_labels, 10)
test_labels_one_hot = tf.keras.utils.to_categorical(test_labels, 10)
```

```python
# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
print(f"Train images shape: {train_images.shape}")
print(f"Test images shape: {test_images.shape}")
print(f"Train labels shape: {train_labels.shape}")
print(f"Test labels shape: {test_labels.shape}")
```

```python
# ----- VISUALIZE SOME IMAGES -----
def visualize_dataset_samples(images, labels, class_names, num_samples=25):
    plt.figure(figsize=(10, 10))
    for i in range(num_samples):
        plt.subplot(5, 5, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i])
        plt.xlabel(class_names[labels[i][0]])
    plt.tight_layout()
    plt.show()

print("\nVisualizing dataset samples...")
visualize_dataset_samples(train_images, train_labels, class_names)
```

```python
# ----- EXPLORATORY DATA ANALYSIS (EDA) ON IMAGE DATASET -----
def perform_eda(images, labels, class_names):
    # Count distribution of classes
    unique_labels, counts = np.unique(labels, return_counts=True)
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(class_names)), counts)
    plt.xticks(range(len(class_names)), class_names, rotation=45)
    plt.title('Class Distribution in CIFAR-10')
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.tight_layout()
    plt.show()

    # Calculate average color distribution
    avg_color_per_class = []
    for i in range(10):
        class_images = images[labels.flatten() == i]
        avg_color = np.mean(class_images, axis=(0, 1, 2))
        avg_color_per_class.append(avg_color)
    
    avg_color_per_class = np.array(avg_color_per_class)
    
    # Display average color distribution
    plt.figure(figsize=(12, 6))
    for i, color in enumerate(['Red', 'Green', 'Blue']):
        plt.subplot(1, 3, i+1)
        plt.bar(range(len(class_names)), avg_color_per_class[:, i])
        plt.xticks(range(len(class_names)), class_names, rotation=45)
        plt.title(f'Average {color} Channel Value per Class')
        plt.xlabel('Class')
        plt.ylabel('Average Value')
    plt.tight_layout()
    plt.show()
    
    # Calculate brightness and contrast
    brightness = np.mean(images, axis=(1, 2, 3))
    contrast = np.std(images, axis=(1, 2, 3))
    
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.hist(brightness, bins=30)
    plt.title('Brightness Distribution')
    plt.xlabel('Mean Pixel Value')
    plt.ylabel('Count')
    
    plt.subplot(1, 2, 2)
    plt.hist(contrast, bins=30)
    plt.title('Contrast Distribution')
    plt.xlabel('Standard Deviation of Pixel Values')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

print("\nPerforming Exploratory Data Analysis...")
perform_eda(train_images, train_labels, class_names)
```

```python
# ----- IMAGE PREPROCESSING TECHNIQUES -----
def apply_preprocessing_techniques(image):
    """Apply various preprocessing techniques to a single image"""
    # Make a copy of the original image
    processed_images = []
    titles = []
    
    # Original
    processed_images.append(image)
    titles.append('Original')
    
    # Grayscale conversion
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    processed_images.append(gray)
    titles.append('Grayscale')
    
    # Gaussian blur
    blurred = cv2.GaussianBlur(image, (5, 5), 0)
    processed_images.append(blurred)
    titles.append('Gaussian Blur')
    
    # Edge detection (Canny)
    edges = cv2.Canny(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY), 100, 200)
    edges = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
    processed_images.append(edges)
    titles.append('Edge Detection')
    
    # Histogram equalization
    img_yuv = cv2.cvtColor(image, cv2.COLOR_RGB2YUV)
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    hist_eq = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
    processed_images.append(hist_eq)
    titles.append('Histogram Equalization')
    
    # Display results
    plt.figure(figsize=(15, 10))
    for i in range(len(processed_images)):
        plt.subplot(2, 3, i+1)
        if len(processed_images[i].shape) == 2:
            plt.imshow(processed_images[i], cmap='gray')
        else:
            plt.imshow(processed_images[i])
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return processed_images

print("\nDemonstrating Image Preprocessing Techniques...")
sample_idx = np.random.randint(0, len(train_images))
sample_image = (train_images[sample_idx] * 255).astype(np.uint8)
processed_images = apply_preprocessing_techniques(sample_image)
```

```python
# ----- IMAGE SEGMENTATION -----
def perform_image_segmentation(image):
    """Perform simple image segmentation using various techniques"""
    # Convert to uint8 for OpenCV functions
    image_uint8 = (image * 255).astype(np.uint8)
    
    # K-means segmentation
    pixel_values = image_uint8.reshape((-1, 3))
    pixel_values = np.float32(pixel_values)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
    k = 4  # Number of clusters
    _, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    centers = np.uint8(centers)
    segmented_image = centers[labels.flatten()]
    segmented_image = segmented_image.reshape(image_uint8.shape)
    
    # Thresholding-based segmentation
    gray = cv2.cvtColor(image_uint8, cv2.COLOR_RGB2GRAY)
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    
    # Display results
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(image)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(segmented_image)
    plt.title('K-means Segmentation')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(binary, cmap='gray')
    plt.title('Thresholding Segmentation')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

print("\nPerforming Image Segmentation...")
perform_image_segmentation(train_images[sample_idx])
```

```python
# ----- DATA AUGMENTATION FOR TRAINING -----
def visualize_augmentation():
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    # Choose a sample image
    img = train_images[sample_idx].reshape((1,) + train_images[sample_idx].shape)
    
    # Generate and plot augmented images
    plt.figure(figsize=(15, 8))
    plt.subplot(2, 4, 1)
    plt.imshow(train_images[sample_idx])
    plt.title("Original")
    plt.axis('off')
    
    i = 2
    for batch in datagen.flow(img, batch_size=1):
        plt.subplot(2, 4, i)
        plt.imshow(batch[0])
        plt.title(f"Augmented #{i-1}")
        plt.axis('off')
        i += 1
        if i > 8:
            break
    plt.tight_layout()
    plt.show()
    
    return datagen

print("\nVisualizing Data Augmentation...")
datagen = visualize_augmentation()
```

```python
# ----- FEATURE EXTRACTION WITH PCA AND T-SNE -----
def extract_features():
    # Sample a subset of the data for visualization (to avoid long computation times)
    n_samples = 1000
    sample_indices = np.random.choice(train_images.shape[0], n_samples, replace=False)
    
    # Flatten images for PCA and t-SNE
    X_flat = train_images[sample_indices].reshape(n_samples, -1)
    y_flat = train_labels[sample_indices].flatten()
    
    # Apply PCA
    print("Applying PCA...")
    pca = PCA(n_components=50)
    X_pca = pca.fit_transform(X_flat)
    
    # Plot PCA variance explained
    plt.figure(figsize=(10, 6))
    explained_variance = pca.explained_variance_ratio_
    plt.plot(np.cumsum(explained_variance))
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('PCA Explained Variance')
    plt.grid(True)
    plt.show()
    
    # Apply t-SNE on PCA results (for efficiency)
    print("Applying t-SNE (this may take a few minutes)...")
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(X_pca)
    
    # Plot t-SNE results
    plt.figure(figsize=(12, 10))
    for i, cls in enumerate(class_names):
        idx = y_flat == i
        plt.scatter(X_tsne[idx, 0], X_tsne[idx, 1], label=cls, alpha=0.6)
    plt.title('t-SNE Visualization of CIFAR-10')
    plt.legend()
    plt.show()

print("\nExtracting Features with PCA and t-SNE...")
extract_features()
```

```python
# ----- BUILD A CONVOLUTIONAL NEURAL NETWORK (CNN) MODEL -----
def build_cnn_model():
    model = models.Sequential([
        # First convolutional block
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),
        
        # Second convolutional block
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.3),
        
        # Third convolutional block
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.4),
        
        # Fully connected layers
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    
    return model

print("\nBuilding CNN Model...")
cnn_model = build_cnn_model()
cnn_model.summary()
```

```python
def compile_model(model):
    # Compile with Adam optimizer
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

print("\nCompiling the Model...")
cnn_model = compile_model(cnn_model)
```

```python
# ----- TRAIN THE MODEL -----
def train_model(model, train_images, train_labels_one_hot, test_images, test_labels_one_hot):
    # Use data augmentation for training
    datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        zoom_range=0.1
    )
    
    # Create callback for early stopping
    early_stop = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )
    
    # Create checkpoint callback
    checkpoint_path = "cifar10_cnn_model_checkpoint.h5"
    checkpoint = tf.keras.callbacks.ModelCheckpoint(
        checkpoint_path,
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=False,
        verbose=1
    )
    
    # Train the model
    print("\nTraining the Model (this may take some time)...")
    history = model.fit(
        datagen.flow(train_images, train_labels_one_hot, batch_size=64),
        epochs=30,
        validation_data=(test_images, test_labels_one_hot),
        callbacks=[early_stop, checkpoint]
    )
    
    return model, history

print("\nTraining the Model...")
cnn_model, history = train_model(cnn_model, train_images, train_labels_one_hot, test_images, test_labels_one_hot)
```

```python
# ----- EVALUATE THE MODEL -----
def evaluate_model(model, test_images, test_labels, test_labels_one_hot, class_names):
    # Evaluate the model
    print("Evaluating the Model...")
    test_loss, test_acc = model.evaluate(test_images, test_labels_one_hot, verbose=2)
    print(f"Test accuracy: {test_acc:.4f}")
    
    # Predict on test data
    predictions = model.predict(test_images)
    pred_classes = np.argmax(predictions, axis=1)
    true_classes = test_labels.flatten()
    
    # Create confusion matrix
    cm = confusion_matrix(true_classes, pred_classes)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(true_classes, pred_classes, target_names=class_names))

print("\nEvaluating the Model...")
evaluate_model(cnn_model, test_images, test_labels, test_labels_one_hot, class_names)
```

```python
# ----- PLOT TRAINING HISTORY (LOSS & ACCURACY) -----
def plot_training_history(history):
    plt.figure(figsize=(12, 5))
    
    # Plot training & validation accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='lower right')
    
    # Plot training & validation loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper right')
    
    plt.tight_layout()
    plt.show()

print("\nPlotting Training History...")
plot_training_history(history)
```
